```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
z = x ** 3
```

```python
y
```

```python
z
```

```python
plt.plot(x,y)
plt.xlabel('X_Label')
plt.ylabel('Y_Label')
plt.title('Title')
```

```python
plt.plot(x,y,label='X squared')

# Label
plt.xlabel('X_Label')
plt.ylabel('Y_Label')

##Adding Title
plt.title('Title')
plt.legend()
plt.show()
```

```python
plt.plot(x,z,label='X cubed', color = 'orange')
plt.legend(loc = 0) ##Best position as determined by matplotlib
plt.title('X cubed')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color = 'r')
plt.title('X squared')

plt.subplot(1,2,2)
plt.plot(x,z,color = 'b')
plt.title('X cubed')
```

```python
##Above we used the functional method of plotting
##Now we will use Object Oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_title('X squared')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4, 0.2])

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
fig,axes = plt.subplots(1,3) ##Using manually in individual plots
axes[0].plot(x,x**1, color = 'red')
axes[1].plot(x,x**2, color = 'blue')
axes[2].plot(x,x**3, color = 'blue')


for ax in axes: ##Using in for loop
    ax.set_title('Title')
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')

fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X squared', color = 'red')
ax.plot(x,x**3, label = 'X cubed', color = 'green')
ax.legend()
```
