```python
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips['size'].nunique()
```

```python
tips['size'].unique()
```

```python
tips.groupby('size').count()
```

```python
tips.groupby('sex').mean()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
sns.histplot(data = tips, x = 'total_bill')
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex')
```

```python
sns.pairplot(data = tips)
```

```python
sns.rugplot(x = 'total_bill', data = tips)
```

```python
# Categorical plots
sns.boxplot(x = 'sex', y = 'total_bill', data = tips, hue = 'day')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.countplot(x = 'day', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
#Should not be used in case of big data
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
#Scatterplot between a categorical variable and a numerical variable
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', jitter = True)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box')
```
